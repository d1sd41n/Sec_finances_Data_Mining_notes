```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
endpoint = "https://www.sec.gov/include/ticker.txt"
#get all tikers and ciks
response = requests.get(url = endpoint)

# print status code
print(response.status_code)
print(response.url)
```

```python
cik = response.text.split("\n")
len(cik)
```

```python
for c, i in enumerate(cik):
    cik[c] = i.split("\t")
print(cik)
```
