Documents links finder

With this code, all the links of each document of a specific company stored in the database will be searched, and will be printed on the screen

```python
# import our libraries
import requests
import urllib
from bs4 import BeautifulSoup
```

```python
# url base
base_url = r"https://www.sec.gov/Archives/edgar/data"

cik_num = '/886982/' # company CIK

filings_url = base_url + cik_num + "/index.json" 

# get the json of all folders of all reports
content = requests.get(filings_url)
decoded_content = content.json()

# get multiple filings
for filing in decoded_content["directory"]["item"]:
    # defining each filing number
    filing_num = filing["name"]
    # url of a json report
    filing_url = base_url + cik_num + filing_num + "/index.json"
    
    content = requests.get(filing_url)
    document_content = content.json()
    
    #iterating files of a report
    for document in document_content['directory']["item"]:
        doc_name = document["name"]
        document_url = base_url + cik_num + filing_num + "/" + doc_name
        print(document_url)
```
